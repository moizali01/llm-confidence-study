## Science MCQ Dataset Sample (Kaggle)

### Importing Dataset

```python
import kagglehub
import os
import pandas as pd

path = kagglehub.dataset_download("lizhecheng/llm-science-dataset")
dataset_path = os.path.join(path, 'High Quality Dataset.csv')
df = pd.read_csv(dataset_path)

df
```

### Cleaning Dataset

EXCLUDE ID column.

```python
#drop first column
df = df.drop(df.columns[0], axis=1)
df
```

### Adjusting and Saving Dataset

Convert the dataset to a format that can be used in the model.

```python
prompts = []
answers = []
options = ['A', 'B', 'C', 'D', 'E']
count = 0

for i, each in df.iterrows():

    if count == 5000:
        break

    prompt = f"""Question: {each['prompt']}

A) {each['A']}
B) {each['B']}
C) {each['C']}
D) {each['D']}
E) {each['E']}

Choose the best option."""
    
    print(prompt)

    prompts.append(prompt)
    answers.append(each['answer'])
    count+=1


new_df = pd.DataFrame({'prompt': prompts, 'answer': answers})

new_df.to_csv('ScienceMCQ_5000_sample.csv', index=False)
```
