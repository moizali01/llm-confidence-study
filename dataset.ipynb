### MMLU Dataset Sample (Kaggle)

```python
import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("peiyuanliu2001/mmlu-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd

# Load the CSV file (replace 'filename.csv' with the actual file name)
dataset_path = os.path.join(path, 'test.csv')
df = pd.read_csv(dataset_path)

# View the first few rows of the dataset
df
```

### Create Prompt From the Dataset

```python
prompts = []
answers = []
count = 0

for i, each in df.iterrows():

    if count == 5000:
        break

    prompt = f"""Question: {each['prompt']}

A) {each['A']}
B) {each['B']}
C) {each['C']}
D) {each['D']}

Choose the best option."""
    
    print(prompt)

    prompts.append(prompt)
    answers.append(each['answer'])
    count+=1


new_df = pd.DataFrame({'prompt': prompts, 'answer': answers})

new_df.to_csv("MMLU_5000_sample.csv", index=False)
```
